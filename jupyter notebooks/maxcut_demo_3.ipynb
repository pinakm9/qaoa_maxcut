```python
# add modules to Python's search path
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

# import our weighted graph and max-cut modules
from modules import wgraph as gr
from modules import maxcut as mc
```

## A 3-triangle graph

```python
edges = [(0, 1), (1, 2), (2, 3), (3, 4), (0, 3), (2, 4), (0, 2)]
g = gr.gen_wgraph_from_edges(edges=edges, unweighted=False, max_weight=10.0)
g.draw()
```

## Solve max-cut using QAOA

```python
mxc = mc.MaxCut(g)
mxc.build_objective(num_shots=1729) # the fabled taxi-cab number makes an appearance
mxc.solve(depth=10, max_iter=3000)
```
